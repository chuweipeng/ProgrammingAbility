```python
#Sentiment Analysis
import numpy
import nltk
import csv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('punkt')
nltk.download('vader_lexicon')
```

```python
import matplotlib.pyplot as plt
sid = SentimentIntensityAnalyzer()

sentiment = []

for t in federalist["text"]:
  scores = sid.polarity_scores(t)
  sentiment.append(scores['compound'])
```

```python
federalist['sentiment'] = sentiment
federalist
```
